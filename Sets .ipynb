### Basic Properties

```python
# 1. Lenght : total unique elements

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )

print( len(s1) )
```

```python
# 2. Indexingg and Slicing

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )


print( s1[4 : 10] )
```

```python
# 3. Concat and Repeat
s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )

# print( s1 * 6 )
# print( s1 + s1 )
```

```python
# 4. Membership (in)

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )

print( 1 in s1 )
print( 10 in s1 )
```

```python
s1 = [ 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  ]

s1 = list( set(s1) )
s1
```

```python
## No replacement

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )


s1[2] = 99
```

```python
# add : insert single elements


s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )

s1.add(99)
print(s1)

# s1.add([99, 55])
# print(s1)

s1.add(99)
print(s1)
```

```python
# update : insert multiple elements

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print( s1 )

s1.update([12, 25, 'python', 10])
s1
```

```python
# pop : ALWAYS remove element at any POSITION

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print(s1)

print( s1.pop() )
print(s1)


print( s1.pop() )
print(s1)

print( s1.pop() )
print(s1)

print( s1.pop() )
print(s1)

```

```python
# remove / discard : remove element by VALUE

s1 = { 10, 'python', False, 3.5, 10 , 10, 'python', 10, 10 , False, 'python', -6, 2j  }
print(s1)

s1.discard(False)
print(s1)


s1.remove(10)
print(s1)
```

```python
# 'issubset', 'issuperset',  

a = {10, 20, 10, 20, 30, 30, 10}
b = {10, 30}
c = {10, 'python'}

print(a)
print(b)
print(c)
```

```python
print( a.issuperset(b) )

print( b.issubset(a) )
```

```python
print( a.issuperset(c) )

print( c.issubset(a) )
```

```python

# , 'difference',
# ' 'intersection',  'isdisjoint', 

# ' 'union', 



a = { 10, 'python', 3.5, False, 10, False }
b = { 'python', "sql", -5, '@', -5 }

```

```python
union_ = a.union(b)
union_
```

```python
inter_ = a.intersection(b)
inter_
```

```python
symm_diff = a.symmetric_difference(b)  # Exclusive OR
symm_diff
```

```python
a_ = a.difference(b)  # a = a - ( a intersection b )
a_
```

```python
b_ = b.difference( a )
b_
```
